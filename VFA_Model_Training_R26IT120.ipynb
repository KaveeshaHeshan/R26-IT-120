### Remove Outliers (IQR)

```python
# Remove Outliers (IQR Method)
def remove_outliers(df, col):
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return df[
        (df[col] >= Q1 - 1.5*IQR) &
        (df[col] <= Q3 + 1.5*IQR)
    ]

print(f"Before outlier removal: {len(latex_vfa_df)} rows")

for col in ['pH', 'turbidity_ntu', 'temperature_c', 'vfa']:
    before = len(latex_vfa_df)
    latex_vfa_df = remove_outliers(latex_vfa_df, col)
    removed = before - len(latex_vfa_df)
    print(f"{col}: {removed} outliers removed")

print(f"\nAfter outlier removal: {len(latex_vfa_df)} rows")
print(f"Grade distribution:\n{latex_vfa_df['grade'].value_counts()}")
```

### Random Forest Model Training + GridSearchCV

```python
from sklearn.ensemble        import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators':      [50, 100, 200],
    'max_depth':         [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'max_features':      ['sqrt', 'log2', None],
}

rf = RandomForestRegressor(random_state=42, n_jobs=-1)

grid_search = GridSearchCV(
    rf, param_grid,
    cv=5,
    scoring='neg_mean_absolute_error',
    n_jobs=-1,
    verbose=1
)

print(" Training started... (5-10 minutes)")

grid_search.fit(
    X_train_scaled, y_train,
    sample_weight=sample_weights
)

best_model = grid_search.best_estimator_

print(f"\n Training complete!")
print(f"   Best params: {grid_search.best_params_}")
print(f"   Best CV MAE: {-grid_search.best_score_:.5f}")
```

```python


# ── Grade assign function
def assign_grade(vfa):
    if vfa < 0.05:  return 'A'
    if vfa < 0.08:  return 'B'
    return 'C'

#  Build table
sample_indices = np.random.choice(
    len(y_test), size=20, replace=False
)

actual_vals    = y_test.values[sample_indices]
predicted_vals = y_pred[sample_indices]
errors         = np.abs(actual_vals - predicted_vals)

actual_grades    = [assign_grade(v) for v in actual_vals]
predicted_grades = [assign_grade(v) for v in predicted_vals]
grade_match      = [
    '' if a == p else ''
    for a, p in zip(actual_grades, predicted_grades)
]

sample_df = pd.DataFrame({
    'Sample'          : [f'S{i+1:02d}' for i in range(20)],
    'Actual VFA'      : [f'{v:.5f}' for v in actual_vals],
    'Predicted VFA'   : [f'{v:.5f}' for v in predicted_vals],
    'Error (MAE)'     : [f'{e:.5f}' for e in errors],
    'Actual Grade'    : actual_grades,
    'Predicted Grade' : predicted_grades,
    'Match'           : grade_match,
})

print("── SAMPLE PREDICTIONS TABLE (20 samples) ──\n")
print(sample_df.to_string(index=False))

# Stats (fixed)
correct = sum(1 for m in grade_match if m == '')
total   = len(grade_match)

print(f"\nGrade Accuracy : {correct}/{total} ({correct/total*100:.1f}%)")
print(f"Max Error      : {errors.max():.5f}")
print(f"Min Error      : {errors.min():.5f}")
print(f"Avg Error      : {errors.mean():.5f}")

#  Grade Accuracy on full test set
all_actual_grades    = [assign_grade(v) for v in y_test.values]
all_predicted_grades = [assign_grade(v) for v in y_pred]

full_correct = sum(
    1 for a, p in zip(all_actual_grades, all_predicted_grades)
    if a == p
)
full_total = len(all_actual_grades)

print(f"\n── FULL TEST SET GRADE ACCURACY ──")
print(f"Correct  : {full_correct}/{full_total}")
print(f"Accuracy : {full_correct/full_total*100:.2f}%")
```
