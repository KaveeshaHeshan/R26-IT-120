## Data Preprocessing

### Check Missing Values

```python
print("Missing Values:")
print(latex_vfa_df.isnull().sum())
```

### Fix Missing Values

```python
# Check Missing Values
print("=== Missing Values ===")
print(latex_vfa_df.isnull().sum())
print("\nTotal Missing:", latex_vfa_df.isnull().sum().sum())
print("Missing %:")
print((latex_vfa_df.isnull().sum() / len(latex_vfa_df) * 100).round(2))
```

```python
num_cols = ['pH', 'turbidity_ntu', 'turbidity_log',
            'temperature_c', 'ammonia_content']

latex_vfa_df[num_cols] = latex_vfa_df[num_cols].fillna(
    latex_vfa_df[num_cols].mean()
)

# Check කරන්න
print("Missing Values After Fix:")
print(latex_vfa_df.isnull().sum().sum())
```

### Remove Unnecessary Columns

```python
# Remove unnecessary columns
latex_vfa_df = latex_vfa_df.drop(columns=[
    'pH_duplicate',
    'turbidity_ntu',
    'drc'
])

print("Remaining columns:", latex_vfa_df.columns.tolist())
print("Shape:", latex_vfa_df.shape)
```

### Fix Missing Values

```python
print("Before fix:")
print(latex_vfa_df[['pH','turbidity_ntu','temperature_c']].isnull().sum())

# Median imputation
for col in ['pH', 'turbidity_ntu', 'temperature_c']:
    median_val = latex_vfa_df[col].median()
    missing    = latex_vfa_df[col].isnull().sum()
    latex_vfa_df[col] = latex_vfa_df[col].fillna(median_val)
    print(f"{col}: {missing} missing → filled with median ({median_val:.4f})")

# Verify
print(f"\nAfter fix:")
print(latex_vfa_df[['pH','turbidity_ntu','temperature_c']].isnull().sum())
print(f"\nTotal missing: {latex_vfa_df.isnull().sum().sum()}")
print(f"All missing values fixed!")
```
