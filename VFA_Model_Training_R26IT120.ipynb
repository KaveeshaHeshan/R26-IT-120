### Remove Outliers (IQR)

```python
# Remove Outliers (IQR Method)
def remove_outliers(df, col):
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return df[
        (df[col] >= Q1 - 1.5*IQR) &
        (df[col] <= Q3 + 1.5*IQR)
    ]

print(f"Before outlier removal: {len(latex_vfa_df)} rows")

for col in ['pH', 'turbidity_ntu', 'temperature_c', 'vfa']:
    before = len(latex_vfa_df)
    latex_vfa_df = remove_outliers(latex_vfa_df, col)
    removed = before - len(latex_vfa_df)
    print(f"{col}: {removed} outliers removed")

print(f"\nAfter outlier removal: {len(latex_vfa_df)} rows")
print(f"Grade distribution:\n{latex_vfa_df['grade'].value_counts()}")
```

### Check Class Balance

```python
# Check Class Balance
import pandas as pd

# Grade counts
grade_counts = pd.Series(y_train).apply(
    lambda v: 'A' if v < 0.05 else ('B' if v < 0.08 else 'C')
).value_counts()

total = len(y_train)

print("Train set grade distribution:")
print(f"{'Grade':<10} {'Count':<10} {'%':<10}")
print("-" * 30)
for grade in ['A', 'B', 'C']:
    count = grade_counts.get(grade, 0)
    pct   = (count / total) * 100
    print(f"{grade:<10} {count:<10} {pct:.1f}%")

print(f"\nTotal train samples: {total}")

# Balance check
max_count = grade_counts.max()
min_count = grade_counts.min()
ratio     = max_count / min_count

print(f"\nImbalance ratio: {ratio:.2f}")
if ratio < 1.5:
    print("Balanced — sample weights still recommended")
else:
    print("Imbalanced — sample weights required")
```
